# Scatter Plot
digunakan untuk melihat apakah ada korelasi antara dua set data

misalkan kita punya data berat badan dan tinggi badan anak sekolah 100 orang. kita mau tau apakah anak yang semakin tinggi, maka berat badan semakin berat juga? apakah ada korelasi antara tinggi dan berat badan? kita gunakan scatter plot. di statistik kita akan belajar untuk cari correlation coefficient r - sebuah angka yang menunjukkan seberapa terkorelasinya, positif/ negatif korelasinya.

```python
# apakah ada korelasi antara jumlah imigran Korea Selatan dan Japan?
"Republic of Korea" in df.index
```

```python
"Japan" in df.index
```

```python
df_jpkor = df.loc[["Japan","Republic of Korea"], years].T
df_jpkor.head()
```

```python
df_jpkor.plot(kind="scatter", x="Japan", y="Republic of Korea")
```

```python
df_jpkor.plot(kind="scatter", x="Japan", y="Republic of Korea", figsize=(12,6))

plt.title("Korelasi jumlah imigran Korea Selatan dan Japan")
```

ada sedikit korelasi positif antara jumlah imigran Korea Selatan dan Japan

```python
# apakah ada korelasi antara tahun dengan jumlah imigran dari Japan
# index nya kujadikan kolom "year" dulu
df_jpkoryear = df_jpkor.reset_index()
df_jpkoryear = df_jpkoryear.rename(columns={"index":"year"})
df_jpkoryear.head()
```

```python
df_jpkoryear.plot(kind="scatter", x="year", y="Japan")
```

```python
# dari tahun ke tahun Japan makin banyak kirim imigran ke Canada
```

```python
df_jpkoryear.plot(kind="scatter", x="year", y="Republic of Korea")
```

```python
# membuat figure untuk menggabungkan beberapa plot di satu image

fig = plt.figure()
ax0 = fig.add_subplot(1, 2, 1) 
ax1 = fig.add_subplot(1, 2, 2) 

df_jpkoryear.plot(figsize=(12,6), kind="scatter", x="year", y="Japan", ax=ax0)

df_jpkoryear.plot(figsize=(12,6), kind="scatter", x="year", y="Republic of Korea", ax=ax1)

fig.suptitle("Korelasi Imigrasi Japan dan Korea dengan Tahun")
```

```python
# menggabungkan beberapa scatter plot di plot yang sama

plt.figure(figsize=(12,6))
plt.scatter(df_jpkoryear['year'], df_jpkoryear['Japan'])
plt.scatter(df_jpkoryear['year'], df_jpkoryear['Republic of Korea'])

plt.title("Gabung di satu plot")
```
