## DataFrame 
### Membuat
- buat dari beberapa Series atau beberapa List
- import data excel, csv, table,... 2D data

### Membaca
- info() melihat informasi tentang DataFrame, berapa baris, berapa kolom, setiap kolom tipe data apa
- describe() melihat informasi statistik untuk kolom2 data di DataFrame yang bersifat numerik
- head() dan tail() untuk melihat beberapa baris pertama dan baris terakhir
- indexing dan slicing 
    - baca 1 baris/ bbrp baris data tertentu .loc() dan .iloc()
    - baca 1 kolom/ bbrp kolom data tertentu
    - kombinasi 1 data di baris dan kolom tertentu atau bbrp baris dan bbrp kolom tertentu
- .columns melihat judul2 kolom DataFrame
- .index melihat judul2 index DataFrame
- query - mengambil data yang memenuhi kondisi terentu
- groupby() digunakan untuk merangkum datanya dari kolom tertentu lalu kita agregasi (sum, mean, max, min, median, std, ...)
- value_counts() untuk menghitung data di satu kolom yang unik apa saja, dan masing2 muncul berapa kali
- nunique() untuk menghitung ada berapa data unik di sebuah kolom

### Merubah, edit
- ubah judul kolom
- tambah kolom
- hapus kolom
- ubah judul baris
- tambah baris
- hapus baris
- membuat kolom tertentu menjadi kolom index (judul baris)
- membuat kolom index menjadi data di 1 kolom baru
- dropna() menghapus baris atau kolom data yang ada missing value
- fillna() mengisi data2 yang kosong, missing values di DataFrame kita langsung isi semua

### Menghapus
- hapus baris/kolom/ missing values
