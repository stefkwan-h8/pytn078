```python
# load data tindak kriminal di San Francisco
# data ini ada kolom untuk latitude dan longitude
# data lat dan long nya bisa kita pakai untuk bikin marker di map

# jika di colab, download dulu datanya pakai wget
# !wget --quiet dataset/Police_Department_Incidents_-_Previous_Year__2016_.csv -O incidents.csv
# data_sf = pd.read_csv("incidents.csv")

data_sf = pd.read_csv("dataset/Police_Department_Incidents_-_Previous_Year__2016_.csv")
data_sf.info()
```

```python
# kita akan pakai 100 titik untuk marking di map tindak kriminal yang terjadi di SF
df100 = data_sf.head(100)
df100.head()
```

```python
# peta SF
latitude = 37.77
longitude = -122.42
map_sf = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
# kita gunakan iterrows() untuk melewati setiap baris data satu per satu
# kita akan pakai lokasi lat dan long serta penjelasan Category untuk membuat 100 marker

for (i, crime) in df100.iterrows():
    # print(crime["X"], crime["Y"], crime["Category"])
    # print()
    crime_marker = folium.Marker(location=[crime["Y"], crime["X"]], popup=crime["Category"])
    map_sf.add_child(crime_marker)
    
```

```python
map_sf
```

```python
# sedikit lebih rapi kalau marker2 nya kita kumpulin jadi satu "feature group" baru kita masukin ke map
map_sf = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = folium.map.FeatureGroup()
for (i, crime) in df100.iterrows():
    # print(crime["X"], crime["Y"], crime["Category"])
    # print()
    crime_marker = folium.Marker(location=[crime["Y"], crime["X"]], popup=crime["Category"])
    incidents.add_child(crime_marker)

map_sf.add_child(incidents)
```

```python
# ada juga tipe marker bulat, CircleMarker
map_sf = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = folium.map.FeatureGroup()
for (i, crime) in df100.iterrows():
    # print(crime["X"], crime["Y"], crime["Category"])
    # print()
    crime_marker = folium.CircleMarker(
                            location=[crime["Y"], crime["X"]], 
                            popup=crime["Category"],
                            radius=5, # define how big you want the circle markers to be
                            color='yellow',
                            fill=True,
                            fill_color='blue',
                            fill_opacity=0.6)
    incidents.add_child(crime_marker)

map_sf.add_child(incidents)
```

```python
# pakai plugin MarkerCluster

map_sf = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = plugins.MarkerCluster()
# incidents = folium.map.FeatureGroup()

for (i, crime) in df100.iterrows():
    # print(crime["X"], crime["Y"], crime["Category"])
    # print()
    crime_marker = folium.CircleMarker(
                            location=[crime["Y"], crime["X"]], 
                            popup=crime["Category"],
                            radius=5, # define how big you want the circle markers to be
                            color='yellow',
                            fill=True,
                            fill_color='blue',
                            fill_opacity=0.6)
    incidents.add_child(crime_marker)

map_sf.add_child(incidents)
```

setelah maps nya kita "kosongin" secara visual, kita isi lagi, kita gambar isinya

disini kita akan gunakan data imigrasi Canada, untuk warnain negara yang makin banyak kirim imigran warna nya makin gelap, yang tidak/sedikit kirim imigran ke Canada warnanya makin tipis

kita perlu:
1. data jumlah imigran dari setiap negara ke Canada --> canada.xlsx
2. data border negara di dunia dalam bentuk kumpulan angka (lat long) --> wordl geo .json file

```python
df_can = pd.read_excel("dataset/Canada.xlsx", 
                        sheet_name="Canada by Citizenship",
                        skiprows=20,
                        skipfooter=2)

df = df_can.rename(columns={"OdName":"Country", "AreaName":"Area", "RegName":"Region", "DevName":"Dev"})
df = df.drop(columns=['Type', 'Coverage', 'AREA', 'REG', 'DEV'])
df["Total"] = df.sum(axis=1)
display(df.head(2))

# !wget --quiet https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/world_countries.json -O world_countries.json
# print('GeoJSON file downloaded!')

world_geo = r'dataset/world_countries.json'
```

```python
world_map = folium.Map(
   location=[0,0], 
   zoom_start=2, 
   tiles='"http://{s}.tiles.mapbox.com/v4/wtgeographer.2fb7fc73/{z}/{x}/{y}.png?access_token=pk.eyJ1Ijoic3RlZmt3YW4iLCJhIjoiY2twMjV2ajdkMTZrbzJucHI3YTgxbjM2diJ9.oHD6Rw0KJYhnaYzK2pL__Q',
   attr='XXX Mapbox Attribution'
)

fc = folium.Choropleth(
        geo_data=world_geo,
        data=df,
        columns=['Country', 'Total'],
        key_on='feature.properties.name',
        fill_color='YlOrRd', #Yellow, Orange, Red
        fill_opacity=0.7, 
        line_opacity=0.2,
        legend_name='Immigration to Canada')

world_map.add_child(fc)
```
